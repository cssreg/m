```python
#多线程启动，--api指/docs里有i2p和p2p等，--share值生产gradio.live链接
from threading import Thread
%cd /kaggle/working/stable-diffusion-webui
def RunWebUi():
    !/opt/conda/envs/webui_env/bin/python3 launch.py --port=8000 --ui-config-file=/kaggle/working/stable-diffusion-webui/ui-config.json --ui-settings-file=/kaggle/working/stable-diffusion-webui/config.json --xformers --api --enable-insecure-extension-access
p = Thread(target = RunWebUi)
import inspect
import ctypes
def _async_raise(tid,exctype):
    tid = ctypes.c_long(tid)
    if not inspect.isclass(exctype):
        exctype = type(exctype)
    res = ctypes.pythonapi.PyThreadState_SetAsyncExc(tid,ctypes.py_object(exctypes))
    if res == 0:
        raise ValueError("invalid thread id")
    elif res != 1:
        ctypes.pythonapi.PyThreadState_SetAsyncExc(tid,None)
        raise SystemError("PyThreadState_SetAsyncExc failed")
def stop_thread(thread):
    _async_raise(thread.ident,SystemExit)
```
